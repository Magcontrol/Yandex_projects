### Шаг 3.1 Посмотрим, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

После построения нескольких графиков пришел к выводу, что проще написать функцию

```python
def sozdanie_graf(data, groupby_column, func, nazvanie='name'):
    plt.style.use('seaborn-deep')
    df_to_plot = data.groupby(groupby_column)[nazvanie]
    if func == 'count':
        data_graf = df_to_plot.count()
        figsize = (15, 5)
        plot = data_graf.plot(kind='bar', y=nazvanie, figsize=figsize, ec='black')
        plot.set_xlabel(groupby_column)
        plot.set_ylabel(func)
    elif func == 'sum':
        data_graf = df_to_plot.sum().sort_values()
        figsize = (15, 10)
        plot = data_graf.plot(kind='bar', y=nazvanie, figsize=figsize, ec='black')
        plot.set_xlabel(groupby_column)
        plot.set_ylabel(func)
```

```python
sozdanie_graf(data, 'year_of_release', 'count')
plt.xlabel('Год релиза')
plt.ylabel('Количество')
plt.title('График зависимости количества от года релиза')
plt.show()
```

## Шаг 5. Проверим гипотезы.

### Шаг 5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

```python
alpha = .05
xbox_ratings_users = data[(data['platform'] == 'XOne') & (data['user_score'] > 0)]['user_score']
pc_ratings_users = data[(data['platform'] == 'PC') & (data['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг  игр для платформы Xbox', xbox_ratings_users.mean())
print('Средняя пользовательский рейтинг  игр для платформы PC', pc_ratings_users.mean())
results = st.ttest_ind(xbox_ratings_users,pc_ratings_users)
print('p-значение', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

Отвергаем нулевую гипотезу в пользу альтернативной => комьютер игры получают оценки выше нежели xbox

### Шаг 5.2 Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports равны

Н1 (альтернативная гипотеза)- средние пользовательские рейтинги жанров Action и Sports отличаются

```python
action_ratings_users = data[(data['genre'] == 'Action') & (data['user_score'] > 0)]['user_score']
sports_ratings_users = data[(data['genre'] == 'Sports') & (data['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг  игр жанра Экшен', action_ratings_users.mean())
print('Средняя пользовательский рейтинг  игр жанра Спорт', sports_ratings_users.mean())
new_results = st.ttest_ind(action_ratings_users, sports_ratings_users)
print('p-значение', new_results.pvalue)
if new_results.pvalue > alpha:
    print('Не отвергаем нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')

```

Поскольку p-значение 0.0648 больше уровня значимости альфа, мы не можем отвергнуть нулевую гипотезу и можем утверждать, что у нас недостаточно доказательств чтобы предположить значительную разницу в средних пользовательских оценках между жанрами экшен и спорт.
